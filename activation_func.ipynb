### Sigmoid Activation 

1 **Sigmoid transforms the values between the range 0 and 1.**

2 **The sigmoid activation function, also called the logistic function**

3 **It is traditionally a very popular activation function for neural networks.**

4 **The input to the function is transformed into a value between 0.0 and 1.0**

5 **Inputs that are much larger than 1.0 are transformed to the value 1.0**

6 **Sigmoid transforms the values between the range 0 and 1.**

7 **similarly, values much smaller than 0.0 are snapped to 0.0**



```python
''' Sigmoid Function '''
#Sigmoid activation  function will be used for linear classification problem.
#The Value above 0.5 will be considered as 1 & less than 0.5 will be as 0
tf.keras.activations.sigmoid(a).numpy() # Ranges between 0 to 1
```

### Tanh -- Hyperbolic Tangent Function

1 **outputs values between -1.0 and 1.0.**



```python
''' Tanh Function '''
#This is another activation function used for regression problem
tf.keras.activations.tanh(a).numpy() # Ranges between -1 to +1
```

### Relu -- Rectified Linear Unit

**Range: [ 0 to infinity)**

**for all positive values, and zero for all negative values. ---- max(0,z)**

**It’s sparsely activated.**

**a sparse network is faster than a dense network, as there are fewer things to compute.**



```python
''' Relu Function '''
tf.keras.activations.relu(a).numpy() # will give o/p of 0 or value, which so ever is Max
```

```python
tf.keras.activations.relu([20,30,9,8,-9,-3,8]).numpy()
```

## Leaky Relu

**Dying ReLU-- The downside for being zero for all negative values is a problem called “Dying ReLU”**

**A ReLU neuron is “dead” if it’s stuck in the negative side and always outputs 0.once a neuron gets negative, it’s unlikely for it to recover.**


**Leaky ReLU has a small slope for negative values, instead of altogether zero**

**y = ax when x < 0., x = 0.01,**

```python
layer = tf.keras.layers.LeakyReLU()
output = layer([20,30,9,8,-9,-3,8])

list(output.numpy())
```

### Softmax Activation Function

**is a more generalized logistic activation function which is used for multiclass classification**

**returns a probability distribution over the target classes in a multiclass classification problem**

**Negative inputs will be converted into nonnegative values,because of exponential function.**

**Each input will be in the interval (0,1).**


**As the denominator in each Softmax computation is the same, the values become proportional to each other, which makes sure that together they sum to 1**
